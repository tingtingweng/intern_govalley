## extract info from LinkedIn

```python
link = 'https://www.linkedin.com/jobs/search?keywords=Data%20Scientist&location=United%20States&trk=public_jobs_jobs-search-bar_search-submit&redirect=false&position=1&pageNum=0'
```

```python
import requests
website_raw = requests.get(link).text
```

```python
print(website_raw[0:500])
print(len(website_raw))
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_raw, 'html.parser')
print(soup.prettify())
```

```python
?soup.find
```

```python
results = soup.find(name = 'ul',attrs = {'class':'jobs-search__results-list'})
results
```

```python
title_list = []
titles = results.find_all('h3')
for title in titles:
    title_list.append(title.get_text().strip())
title_list
```

```python
company_list = []
companies = results.findAll('a',{'class':'result-card__subtitle-link job-result-card__subtitle-link'})
companies
for company in companies:
    company_list.append(company.get_text().strip())
company_list
```

```python
location_list = []
locations = results.findAll('span',{'class':'job-result-card__location'})
for location in locations:
    location_list.append(location.get_text().strip())
location_list
```

```python
date_list = []
dates = results.findAll('time')
for date in dates:
    date_list.append(date.get('datetime'))
date_list
```

```python
import pandas as pd
df = pd.DataFrame()
df['title'] = title_list
df['company'] = company_list
df['location'] = location_list
df['date'] = date_list
```
